```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# datos = pd.read_csv('ca4663.tsp.txt',delimiter=' ')
# datos = pd.read_csv('it16862.tsp.txt',delimiter=' ')

datos = pd.read_csv('it16862.tsp.txt',delimiter=' ')
datos = datos.iloc[:,1:]
dim = datos.shape[0]-1

x = datos.iloc[:dim,0].values / 1000
y = datos.iloc[:dim,1].values /1000


pais = np.array([y,x]).T
pais.shape

plt.figure(figsize=(20,10))
plt.scatter(pais[:,0],pais[:,1])
np.savetxt('italia.csv',pais.round(7),delimiter=',',fmt = '%.8f');
```
